Ademas de representarlas, usando librerias como estas:

```python
import numpy as np
from scipy.integrate import odeint
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
def logisol(t, x0, r):
    return 1/(1 + (1/x0 - 1) * np.exp(-r * t))
```

```python
x = np.linspace(0,10)
for r in np.arange(-1, 1,.1):
    plt.plot(x, logisol(x, 0.05, r))
plt.show()
```
